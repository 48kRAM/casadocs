# Parameters

task parameters

vis : string

Name of input visibility file

Example

caltable : string

Name of output gain calibration table

Example

toffset : double = 0

Time offset (sec) between interferometric and WVR data

Example

segsource : bool = True

Do a new coefficient calculation for each source

Example

sourceflag : stringArray

Regard the WVR data for these source(s) as bad and do not produce
corrections for it (requires segsource=True)

Example

tie : stringArray

Prioritise tieing the phase of these sources as well as possible
(requires segsource=True)

Example

nsol : int = 1

Number of solutions for phase correction coefficients (nsol\>1 requires
segsource=False)

Example

disperse : bool = False

Apply correction for dispersion

Example

wvrflag : stringArray

Regard the WVR data for these antenna(s) as bad and replace its data
with interpolated values from neighbouring antennas

Example

statfield : string

Compute the statistics (Phase RMS, Disc) on this field only

Example

statsource : string

Compute the statistics (Phase RMS, Disc) on this source only

Example

smooth : string

Smooth calibration solution on the given timescale

Example

scale : double = 1.

Scale the entire phase correction by this factor

Example

spw : intArray

List of the spectral window IDs for which solutions should be saved into
the caltable

Example

wvrspw : intArray

List of the spectral window IDs from which the WVR data should be taken

Example

reversespw : string

Reverse the sign of the correction for the listed SPWs (only needed for
early ALMA data before Cycle 0)

Example

cont : bool = False

Estimate the continuum (e.g., due to clouds) (experimental)

Example

maxdistm : double = 500.

maximum distance (m) of an antenna used for interpolation for a flagged
antenna

Example

minnumants : int = 2

minimum number of near antennas (up to 3) required for interpolation

Allowed Value(s)

1 2 3

Example

mingoodfrac : double = 0.8

If the fraction of unflagged data for an antenna is below this value (0.
to 1.), the antenna is flagged.

Example

usefieldtab : bool = False

derive the antenna AZ/EL values from the FIELD rather than the POINTING
table

Example

refant : stringArray

use the WVR data from this antenna for calculating the dT/dL parameters
(can give ranked list)

Example

offsetstable : string

(experimental) subtract the temperature offsets in this table from the
WVR measurements before calculating the phase corrections

Example
