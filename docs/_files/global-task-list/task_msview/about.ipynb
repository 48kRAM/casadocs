# Description

task description

The **msview** task will display a MeasurementSet in
raster form.&nbsp;Many display and editing
options are available.&nbsp;Executing the **msview** task will
bring up a display panel&nbsp;window, which can be
re-sized (see below).&nbsp;


If no data file was specified,&nbsp;a Load Data window will also appear. Click on the desired
MeasurementSet and the rendered data should
appear on the display panel.&nbsp;A Data Display Options window
will also appear. It has drop-down&nbsp;subsections for
related options, most of which are self-explanatory.&nbsp;

The state of the **msview** task, i.e. the loaded data and related
display&nbsp;options, can be saved in a 'restore' file
for later use.&nbsp;You can provide the restore
filename on the command line or&nbsp;select it
from the Load Data window.

For more detailed on how to use the **msview** task,
please read the dedicated CASADocs chapter on [2-D Visualization
and Flagging of Visibility Data
(viewer/msview)](https://casa.nrao.edu/casadocs-devel/stable/calibration-and-visibility-data/data-examination-and-editing/2-d-visualization-of-visibility-data-msview).
