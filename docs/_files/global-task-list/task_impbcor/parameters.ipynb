# Parameters

task applycal parameters

imagename : string

Name of the input (CASA, FITS, MIRIAD) image

Example

imagename='ngc5921.im'

pbimage : undefined = ""

Name of the image (CASA, FITS, MIRIAD) of the primary beam pattern or an
array of pixel values. Default: ''

Example

outfile : string

Name of output CASA image. Default: none. Must be specified.

Example

overwrite : bool = False

If output file is specified, controls if an already existing file by the
same name can be overwritten. Default: True Options: True\|False If
true, the user is not prompted, the file if it exists is automatically
overwritten.

Example

box : string

Rectangular region to select in direction plane. Default: '' (use the
entire direction plane)

Example

region : undefined

Region selection. Default: '' (use the full image)

Example

chans : string

Channels to use. Default: '' (use all channels)

Example

stokes : string

Stokes planes to use. Default: '' (use all Stokes planes)

Example

mask : string

Mask to use. Default: none

Example

mode : string = divide

Divide or multiply the image by the primary beam image. Default:
'divide' Minimal match supported.

Example

cutoff : double = -1.0

Primary beam cutoff. Default: -1.0 (no cutoff) If mode is "d", all
values less than this will be masked. If "m", all values greater will be
masked. Less than 0, no cutoff (default)

Example

stretch : bool = False

Stretch the mask if necessary and possible? Default: False Options:
False\|True

Example
