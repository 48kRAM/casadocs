# Parameters

task parameters

imagename : string

Name of the input image

Example

imagename='ngc5921.im'

outfile : string

Output image name.

Example

factor : intArray

Binning factors for each axis. Use imhead or ia.summary to determine
axis ordering.

Example

region : undefined

Region selection. Default is to use the full image.

Example

box : string

Rectangular region to select in direction plane. Default is to use the
entire direction plane.

Example

chans : string

Channels to use. Default is to use all channels.

Example

stokes : string

Stokes planes to use. Default is to use all Stokes planes. Stokes planes
cannot be rebinned.

Example

mask : string

Mask to use. Default is none.

Example

dropdeg : bool = False

Drop degenerate axes?

Example

overwrite : bool = False

Overwrite the output if it exists? Default False

Example

stretch : bool = False

Stretch the mask if necessary and possible?

Example

crop : bool = True

Remove pixels from the end of an axis to be rebinned if there are not
enough to form an integral bin?

Example
