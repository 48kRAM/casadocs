# Examples

task examples

**plotprofilemap** chooses the size and number of subplots based on that
specified by the user, though blank data in the set may produce
unexpected results. The appearance of the blank data can be changed
using the *plotmasked* function, as shown below.

&nbsp;

The default case for the M100 CSV data:

plotprofilemap(imagename='M100_TP_CO_cube.spw3.image')

![https://raw.githubusercontent.com/8ee9bc833d57e9f01375e4974c2833c2f15b64a8.png](https://raw.githubusercontent.com/8ee9bc833d57e9f01375e4974c2833c2f15b64a8.png)

&nbsp;

Obviously, the inner 4x5 region only, contains data that is not blanked.
A naive solution is to select only those:

plotprofilemap(imagename='M100_TP_CO_cube.spw3.image',numpanels='4,5')

![https://raw.githubusercontent.com/67255e06643ff3f7320e412611835aa4b8624d72.png](https://raw.githubusercontent.com/67255e06643ff3f7320e412611835aa4b8624d72.png)

But the masked data still persists of course, and causes the outer-edges
to be blanked. A solution is to set *plotmasked='none'*:

plotprofilemap(imagename='M100_TP_CO_cube.spw3.image',plotmasked='none')

![https://raw.githubusercontent.com/4b9eaef3de2494f54104e9f6a891ab1407c95730.png](https://raw.githubusercontent.com/4b9eaef3de2494f54104e9f6a891ab1407c95730.png)&nbsp;

Which produces a plot that focuses only on the emission.
