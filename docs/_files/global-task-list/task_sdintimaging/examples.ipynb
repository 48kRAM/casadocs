# Examples

To run sdintimaging with automatic SD-PSF generation, n-sigma stopping
thresholds, a pb-based mask at the 0.3 gain level, and no other
deconvolution masks (interactive=False).&nbsp; Use the helper function
shown below to extract frequency information from the sd cube to supply
as input to sdintimaging.&nbsp; Note that the sdimage cube must contain
per-plane restoring beams.

from sdint_helper import \*  
sdintlib = SDINT_helper()  
sdintlib.setup_cube_params(sdcube='M100_TP')

> 70\]

> 'Direction', 'Stokes', 'Spectral'\]  
> nchan = 70  
> start = 114732899312.0Hz  
> width = -1922516.74324Hz  
> Found 70 per-plane restoring
> beams\
