# Parameters

task parameters

imagename : string

Name of the input image

Example

imagename='ngc5921.im'

region : undefined

Region selection. Default is to use the full image.

Example

region="my.rgn"

box : string

Rectangular region to select in direction plane. Default is to use the
entire direction plane.

Example

box="100,100,200,200"

chans : string

Channels to use. Default is to use all channels.

Example

stokes : string

Stokes planes to use. Default is to use all Stokes planes.

Example

stokes="iq"

mask : string

Mask to use. Default is none.

Example

stretch : bool = False

Stretch the mask if necessary and possible?

Example

function : string = flux density

Aggregate function to use for computing per channel values. Supported
values are "flux density", "mean", "median", "sum". Minimal match
supported.

Example

unit : string = km/s

Unit to use for the spectral flux calculation. Must be conformant with a
typical spectral axis unit.

Example

major : undefined

Major axis of overriding restoring beam. If specified, must be a valid
quantity.

Example

"4arcsec"

minor : undefined

Minor axis of overriding restoring beam. If specified, must be a valid
quantity

Example

"3arcsec"

logfile : string

File which to write details. Default is to not write to a file.

Example

overwrite : bool = False

Overwrite exisitng ouput file if it exists?

Example

overwrite=true
