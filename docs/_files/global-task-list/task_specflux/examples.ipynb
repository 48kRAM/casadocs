box is selected in pixel units, in addition to a mask calculated by the

**Example 2:&nbsp;**

"integrated_line_flux.log". A CASA region "myregion.crtf" is
specified,&nbsp;in addition to a&nbsp;channel range (channels 15 to 25).

values of 11 and 22 arcseconds.&nbsp;

chans="15\~25", unit="km/s", major="11arcsec", minor="22arcsec",
logfile="integrated_line_flux.log", overwrite=True)&nbsp;
