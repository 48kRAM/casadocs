# Examples

task asdmsummary examples

The **asdmsummary** task prints to the CASA log a description of the
content of a Science Data Model (SDM and ASDM are synonymous; the A in
ASDM stands for "ALMA"). For example, to summarize an SDM from the EVLA:

asdmsummary(asdm='10C-119_sb3070258_1.55628.42186299768')

Prints the following information to the CASA logger. Note that the
output from **asdmsummary** may be quite long, depending on the SDM, and
this has been edited for space. Edits are indicated using "...".
