# initweights

task description

This task provides for initialization of the weight information in the
MS. An overview on weights in CASA is provided in the [Data
Weights](https://casa.nrao.edu/casadocs-devel/stable/calibration-and-visibility-data/data-weights)
section of CASAdocs. For ALMA interferometry and EVLA data, it should
not&nbsp;generally be necessary to use this task, as the weight
information&nbsp;should have been initialized properly at fill time
(when the import was performed with CASA v4.2.2 and later).

Several initialization modes are supported via the *wtmode* parameter.
If *wtmode='nyq'* (the default), the SIGMA and WEIGHT will
be&nbsp;initialized according to bandwidth and integration time.
This&nbsp;is the theoretically correct mode for raw normalized
visibilities&nbsp;(e.g., ALMA). For the EVLA, this is correct if
switched-power&nbsp;and bandpass calibration will later be applied. If
*wtmode='sigma'*, WEIGHT will be initialized according to
the&nbsp;existing SIGMA column. If *wtmode='weight'*, WEIGHT_SPECTRUM
will be initialized according&nbsp;to the existing WEIGHT column;
*dowtspec=True* must be specified in&nbsp;this case. If *wtmode='ones'*,
SIGMA and WEIGHT will be initialized with 1.0,&nbsp;globally. This is a
traditional means of initializing weight&nbsp;information, and is
adequate when the integration time and&nbsp;bandwidth are uniform. It is
not recommended for modern&nbsp;instruments (ALMA, EVLA), where variety
in observational setups&nbsp;is common, and properly initialized and
calibrated weights&nbsp;will be used for imaging sensitivity
estimates.  
  
There are two EXPERIMENTAL modes, *wtmode='tsys'* and
*'tinttsys'*.&nbsp;In the modes, SIGMA and WEIGHT will be initialized
according to $T_{sys}$, bandwidth $\Delta\nu$, and
integration time $t_{int}$ (used only in
'*tinttsys*'),&nbsp;i.e.:

-   *tsys* : $weight=\frac{\Delta\nu}{T_{sys}^2}$
-   *tinttsys*: $weight=\frac{\Delta\nu \, t_{int}}{T_{sys}^2}$

These modes use Tsys values to calculate weight as is done in&nbsp;Tsys
calibration. Tsys values are taken from a Tsys calibration&nbsp;table
given as *tsystable*. Selection of gain field
(*gainfield*),&nbsp;interpolation method (*interp*), and spectral window
mapping (*spwmap*)&nbsp;are supported, too.

Available types of interpolation are:

-   Time: '*nearest*', '*linear*', the variation of those with
    '*perobs*',&nbsp;e.g., '*linearperobs*' (enforce obsId boundaries in
    interpolation)
-   Freq: '*nearest*', '*linear*', '*cubic*', '*spline*', and the
    variation&nbsp;of those with '*flag*', e.g., '*linearflag*'
    (with&nbsp;channelized flag).

See the&nbsp;page&nbsp;of **applycal** for details of interpolations.

**WARNING**: If the weight in an MS is initialized with these modes
and&nbsp;Tsys calibration table is applied
with&nbsp;*calwt=True*&nbsp;after that, the&nbsp;weight would be
contaminated by being divided by the square of Tsys&nbsp;twice. !!!
USERS ARE ADVISED TO USE THESE EXPERIMENTAL MODES WITH CARE !!!

For the above *wtmodes*, if *dowtsp=True* (or if the
WEIGHT_SPECTRUM&nbsp;column already exists), the WEIGHT_SPECTRUM column
will be&nbsp;initialized (uniformly in channel in *wtmode='nyq',
'sigma',*&nbsp;'*weight*', and '*ones*'), in a manner consistent with
the&nbsp;disposition of the WEIGHT column. If the
WEIGHT_SPECTRUM&nbsp;column does not exist, *dowtsp=True* will force its
creation.&nbsp;Use of the WEIGHT_SPECTRUM column is only
meaningful&nbsp;for ALMA data which will be calibrated with
channelized&nbsp;Tsys information, or if the weights will become
channelized&nbsp;after calibration, e.g., via averaging over time-
and&nbsp;channel-dependent flagging (**statwt** is a task for
channel-dependent&nbsp;weight estimation from the data itself).&nbsp;

In non-channelized modes (*wtmode='nyq', 'sigma', 'weight',*
and&nbsp;'*ones*') or when *dowtsp=False,*&nbsp;the SIGMA_SPECTRUM
column will be removed&nbsp;from the MS. On the other hand, the
SIGMA_SPECTRUM column is added and&nbsp;initialized in channelized modes
(*wtmode='tsys'* and '*tinttsys*')&nbsp;if *dowtsp=True* or if the
WEIGHT_SPECTRUM already column exists.

Two additional modes are available for managing the spectral&nbsp;weight
info columns; these should be used with extreme care:
If&nbsp;*wtmode='delwtsp'*, the WEIGHT_SPECTRUM column will be deleted
(if&nbsp;it exists). If *wtmode='delsigsp',* the SIGMA_SPECTRUM
column&nbsp;will be deleted (if it exists). Note that creation
of&nbsp;SIGMA_SPECTRUM is not supported via this method.

**NOTE**: This task does not support any prior
selection.&nbsp;Intialization of the weight information must currently
be done globally or not at all. This is to maintain consistency.
