constructor **singledishms**

This is used to construct a singledishms tool instance. The created
instance is just like the default one ('sdms') but physically
independent from it. This is useful when users want to create their own
tool instance inside scripts/modules to avoid possible conflicts that
may happen when using the default tool instance from various places.

Parameters : None

Example

Manual tool construction is done this way: ssd = casac.singledishms()

function **open**

Close the current MeasurementSet and open a new MeasurementSet instead.
The current state of sdms is retained, except for the data selection.

Parameters

ms_name : undefined

New MeasurementSet to be processed

Example

sdms.open('m100_sd.ms')

function **close**

This is used to close sdms tools. Note that the data is written to disk
and detached from sdms tool. This is a synonym for done.

Parameters : None

Example

sdms.close()

function **done**

This is used to close and sdms tools. Note that the data is written to
disk and detached from sdms tool. This is a synonym for close.

Parameters : None

Example

sdms.done()

function **name**

Returns the name of the attached MeasurementSet.

Parameters : None

Example

sdms.name()

baseline function

clip_threshold_sigma : undefined = 3.0
