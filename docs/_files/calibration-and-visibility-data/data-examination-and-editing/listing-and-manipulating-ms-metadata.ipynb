# MS Metadata List/Change

Using vishead

&nbsp;

The **vishead** task is provided to access keyword information in the
MeasurementSet. The default inputs are:

\

## vishead :: List, get, and put metadata in a MeasurementSet  
vis = '' \

## Name of input visibility file  
mode = 'list' \

## options: list, summary, get, put  
listitems = \[\] \

## items to list (\[\] for all)

The *mode = &rsquo;summary&rsquo;* option just gives the same output as
**listobs**.

For *mode = &rsquo;list&rsquo;* the default options are:
*&rsquo;telescope&rsquo;, &rsquo;observer&rsquo;, &rsquo;project&rsquo;,
&rsquo;field&rsquo;, &rsquo;freq_group_name&rsquo;,
&rsquo;spw_name&rsquo;, &rsquo;schedule&rsquo;,
&rsquo;schedule_type&rsquo;, &rsquo;release_date&rsquo;.* To obtain
other options, put&nbsp;*mode = &rsquo;list&rsquo;* and *listitems =
\[\]*; see [vishead task
pages](https://casa.nrao.edu/casadocs-devel/stable/global-task-list/task_vishead)
for a description of these additional options.

You can use *mode=&rsquo;get&rsquo;* to retrieve the values of specific
keywords, and likewise *mode=&rsquo;put&rsquo;* to change them. The
inputs are:

mode&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
=&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 'get'&nbsp;&nbsp;&nbsp; \#&nbsp;
options: list, summary, get, put  
hdkey&nbsp; &nbsp; &nbsp; &nbsp;&nbsp;&nbsp;
=&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; '' &nbsp; &nbsp;&nbsp; \#&nbsp;
keyword to get/put  
hdindex&nbsp;&nbsp; &nbsp; &nbsp;&nbsp;
=&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; ''&nbsp; &nbsp; &nbsp; \#&nbsp;
keyword index to get/put, counting from zero. ==\>all

and

\#&nbsp; vishead :: List, summary, get, and put metadata in a
MeasurementSet  
mode&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
=&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 'put'&nbsp;&nbsp;&nbsp; \#&nbsp;
options: list, summary, get, put  
hdkey&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;
=&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; ''&nbsp;&nbsp;&nbsp;
\#&nbsp; keyword to get/put  
hdindex&nbsp;&nbsp; &nbsp; &nbsp;&nbsp;
=&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; ''&nbsp;&nbsp;&nbsp;
\#&nbsp; keyword index to get/put, counting from zero. ==\>all  
hdvalue &nbsp;&nbsp; &nbsp; &nbsp;
=&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; ''&nbsp;&nbsp;&nbsp;
\#&nbsp; value of hdkey

For example, a common operation is to change the Telescope name (e.g. if
it is unrecognized), e.g.

CASA \<36\>: vishead('ngc5921.demo.ms',mode='get',hdkey='telescope')  
&nbsp; Out\[36\]:  
&nbsp; (array(\['VLA'\],  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; dtype='\|S4'), {})  
  
CASA \<37\>:
vishead('ngc5921.demo.ms',mode='put',hdkey='telescope',hdvalue='JVLA')  
  
CASA \<38\>: vishead('ngc5921.demo.ms',mode='get',hdkey='telescope')  
&nbsp; Out\[38\]:  
&nbsp; (array(\['JVLA'\],  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; dtype='\|S5'), {})
